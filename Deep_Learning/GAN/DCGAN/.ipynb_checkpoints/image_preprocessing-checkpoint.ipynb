```python
# 사람이름으로 된 폴더 안의 이미지들의 이름을 '사람이름' + '[i번째]' + '.jpg' 로 바꿈

import os 
# 이미지 데이터가 있는 폴더 경로
file_path = '/home/csle/Desktop/celeb_dataset/data'
# file_names = '사람 이름' 의 배열
file_names = os.listdir(file_path)

for name in file_names:
    # image_path = '사람 이름' 폴더 경로
    image_path = os.path.join(file_path, name)
    # image_file_names = image_path 폴더 안에 있는 이미지 파일의 배열
    image_file_names = os.listdir(image_path)
    image_file_names = sorted(image_file_names)
    # image_path 폴더 안에 있는 이미지 파일의 개수만큼 반복
    for i in range(len(os.listdir(image_path))):
        # image_path 폴더 안의 모든 이미지 파일의 이름을 '사람이름i.jpg' 로 바꿈
        os.rename(image_path + '/' + image_file_names[i], image_path + '/' + name +'_'+ str(i)+'.jpg')
    
```
